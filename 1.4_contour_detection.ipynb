```python
import os

import numpy as np
import cv2
import matplotlib.pyplot as plt

dir_path = "./data"
```

### Contour detection


```python
poker = cv2.imread(os.path.join(dir_path, "poker.jpg"))
poker = cv2.cvtColor(poker, cv2.COLOR_BGR2RGB)

plt.imshow(poker)
```

```python
poker_gray = cv2.cvtColor(poker, cv2.COLOR_RGB2GRAY)

plt.imshow(poker_gray, cmap="gray")
```

```python
ret, thresh = cv2.threshold(poker_gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap="gray")
```

```python
contours, hierarchy = cv2.findContours(
    thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
)

image_copy = poker.copy()
cv2.drawContours(image_copy, contours, contourIdx=-1, color=(0, 255, 0), thickness=2)
plt.imshow(image_copy)
```

```python
print(f"Number of contours detected: {len(contours)}")
area = cv2.contourArea(contours[0])
print(f"Contour area for the first contour: {area}")
```

```python
all_areas = [cv2.contourArea(contour) for contour in contours]

plt.hist(all_areas, log=True)
```

```python
filtered_contours = [
    contour for contour in contours if cv2.contourArea(contour) > 15000
]

print(f"Number of contours after filtering: {len(filtered_contours)}")

image_copy = poker.copy()
cv2.drawContours(
    image_copy, filtered_contours, contourIdx=-1, color=(0, 255, 0), thickness=2
)
plt.imshow(image_copy)
```

```python
image_copy = poker.copy()
cv2.drawContours(
    image_copy, filtered_contours, contourIdx=0, color=(0, 255, 0), thickness=cv2.FILLED
)
plt.imshow(image_copy)
```

```python
image_copy = poker.copy()
idx = 0
cv2.drawContours(
    image_copy,
    filtered_contours,
    contourIdx=idx,
    color=(0, 255, 0),
    thickness=cv2.FILLED,
)

mask = np.all(image_copy == (0, 255, 0), axis=-1)
plt.imshow(mask)
```

```python
x, y, w, h = cv2.boundingRect(filtered_contours[idx])
print(f"Bounding box for contour {idx}: x={x}, y={y}, w={w}, h={h}")

offset = 20

image_copy = poker.copy()
bounding_box = image_copy[y - offset : y + h + offset, x - offset : x + w + offset]
bounding_box_mask = mask[y - offset : y + h + offset, x - offset : x + w + offset]

plt.imshow(bounding_box)
plt.show()

plt.imshow(bounding_box_mask)
plt.show()
```

```python
bounding_box_copy = bounding_box.copy()
bounding_box_copy[~bounding_box_mask] = (0, 255, 0)
plt.imshow(bounding_box_copy)
```

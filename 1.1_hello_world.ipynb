### If-elif-else


```python
value = 123

if value == 100:
    print("Value is equal to 100")
elif value > 100:
    print("Value is greater than 100")
else:
    print("Value is less than 100")
```

### For-loop


```python
for i in [0, 10, 20, 24, 37, 100]:
    print(f"{i}°C = {i * 1.8 + 32}°F")
```

```python
for i in range(2, 4):
    for j in range(1, 5):
        print(f"{i}^{j} = {pow(i, j)}")
```

### Function


```python
def mix(a, b, alpha):
    return a * (1 - alpha) + b * alpha


print(mix(0, 255, 0))
print(mix(0, 255, 0.5))
print(mix(0, 255, 1))
```

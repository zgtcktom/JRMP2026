### Import

```python

import math
from math import sqrt

print(math.pow(2, 3))
print(sqrt(16))
```

### If-else

```python
value = 123

if value == 100:
    print("Value is equal to 100")
elif value > 100:
    print("Value is greater than 100")
else:
    print("Value is less than 100")
```

### For-loop

```python
for i in range(5):
    print(pow(2, i))
```

### Function

```python
def mix(a, b, alpha):
    return a * (1 - alpha) + b * alpha

print(mix(0, 255, 0))
print(mix(0, 255, 0.5))
print(mix(0, 255, 1))
```

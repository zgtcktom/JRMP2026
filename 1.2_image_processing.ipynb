```python
import os

import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### Listing directory


```python
dir_path = "./data"
print(os.listdir(dir_path))
```

### Loading Images


```python
file_path = os.path.join(dir_path, "hk_harbour.jpg")
img = cv2.imread(file_path)
plt.imshow(img)
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

### Image data manipulation


```python
print(f"Shape of rgb_img: {rgb_img.shape}")
```

```python
height, width, channels = rgb_img.shape
print(f"Height: {height}, Width: {width}, Channels: {channels}")
```

```python
half_height = height // 2

first_half = rgb_img[0:half_height, 0:width, 0:channels]

print(f"Shape of first_half: {first_half.shape}")

plt.imshow(first_half)
```

```python
half_width = width // 2

top_right_quad = rgb_img[0:half_height, half_width:width, 0:channels]

print(f"Shape of top_right_quad: {top_right_quad.shape}")

plt.imshow(top_right_quad)
```

### Color


```python
green_only = rgb_img[:, :, 1]

print(f"Shape of green_only: {green_only.shape}")
plt.imshow(green_only, cmap="Greens")
```

```python
gray_img = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap="gray")
```

#### Magic of cv2.COLOR_RGB2GRAY?



https://en.wikipedia.org/wiki/Grayscale#Colorimetric_(perceptual_luminance-preserving)_conversion_to_grayscale


```python
red_only = rgb_img[:, :, 0]
green_only = rgb_img[:, :, 1]
blue_only = rgb_img[:, :, 2]

gray_img = 0.2126 * red_only + 0.7152 * green_only + 0.0722 * blue_only
gray_img = gray_img.clip(0, 255).astype(np.uint8)
plt.imshow(gray_img, cmap="gray", vmin=0, vmax=255)
```
